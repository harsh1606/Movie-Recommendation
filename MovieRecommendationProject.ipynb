```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('seaborn-whitegrid')
plt.rcParams['figure.figsize'] = (10,4)
%matplotlib inline
from operator import itemgetter
```

```python
#creating ratings data frame
ratings = pd.read_csv("ml-100k/u.data" , sep = '\t' , usecols=range(3) , names=['user_id' , 'movie_id' , 'rating'])
print(ratings.head(),end="\n\n")
print(f"shape of rating dataframe is : {ratings.shape}")
```

```python
#manuplate data about movie for better result
col_names =[ 'movie_id','title','unknown','Action','Adventure','Animation','Childrens','Comedy','Crime','Documentary','Drama','Fantasy','Film-Noir','Horror','Musical','Mystery','Romance','Sci-Fi','Thriller','War','Western',]
cols = [0,1,] + [var for var in range(5,5+19)]
```

```python
#Creating movie dataframe which store all the feature a movie can have
movie = pd.read_csv("ml-100k/u.item",usecols=cols,names=col_names,sep='|',encoding='latin')
print(f"Shape of movie data frame is : {movie.shape}",end="\n\n")
movie.head()
```

```python
#improve ratings dataframe features
ratings =ratings.groupby('movie_id').agg({'user_id' : np.size,'rating' : np.mean})
ratings.columns = ['N_times' , 'avg_rating']
```

```python
ratings.head(n=20)
```

```python
minimum = min(ratings['N_times'])
maximum = max(ratings['N_times'])
ratings['popularity'] = ratings['N_times'].apply(lambda x:(x-minimum)/(maximum-minimum)) #calculate popularity according to N_times and avg_ratings
ratings.head()
```

```python
#merge movie dataframe and ratings dataframe
df = pd.merge(ratings,movie,on='movie_id')
df.head()
```

```python
df.index = df['movie_id']
df.drop('movie_id' , axis=1 , inplace=True)
df.head()
```

```python
from scipy.spatial.distance import cosine
```

```python
def getDistance(movie_id1,movie_id2):
    pop_dis = abs(df.loc[movie_id1,'popularity']-df.loc[movie_id2,'popularity'])
    genre_dis = cosine(
        list(df.loc[movie_id1,'unknown':'Western']),
        list(df.loc[movie_id2,'unknown':'Western'])
       )
    return pop_dis+genre_dis
```

```python
def getNeighbours(movie_id1,k=5):
    distance = [ (movie_id2,getDistance(movie_id1,movie_id2))
    for movie_id2 in df.index  if movie_id2 != movie_id1  ]
    distance.sort(key=itemgetter(1))
    return [ distance[i][0] for i in range(k) ]
```

```python
getNeighbours(50,5)
```

```python
def similarMovie(movie_id,k=5):
    similar_movie = getNeighbours(movie_id,k)
    print(f"{k} Nearest Moivies to  {df.loc[movie_id,'title']}")
    print("Sr.No\tRatings\tMovie Name")
    c = 1
    for movie in similar_movie : 
          print(f"{c}.\t{df.loc[movie,'avg_rating']:.2f}\t{df.loc[movie,'title']}")
          c += 1
```

```python
similarMovie(50,10)
```
